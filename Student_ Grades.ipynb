```python
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

print(df[['Name', 'Average']])
```

```python
# Find the highest score in each subject and the student who achieved it
math_max = df.loc[df['Math'].idxmax()] #whats inside df.loc returns the highest grade in math but its index
#then df.loc finds that student with the same index
science_max = df.loc[df['Science'].idxmax()]
english_max = df.loc[df['English'].idxmax()]
# Clean and readable output
print(f"Highest score in Math: {math_max['Name']} with {math_max['Math']}")
print(f"Highest score in Science: {science_max['Name']} with {science_max['Science']}")
print(f"Highest score in English: {english_max['Name']} with {english_max['English']}")
```

```python
# Sort the students by their overall average score (highest first)
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1) #caculates the average of the 3 subjects, axis=1 means cacualting on rows
top_3_students = df.sort_values(by='Average', ascending=False).head(3) #sort them descending

print("Top 3 students by average score:")
print(top_3_students[['Name', 'Average']])
```

```python
# List students who scored more than 80 in Math
students_above_80_math = df[df['Math'] > 80]#generates a new dataframe with only the students who scored higher than 80

print("Students who scored above 80 in Math:")
print(students_above_80_math[['Name', 'Math']])
```

```python
# Calculate the total score (sum of Math, Science, and English) for each student
df['Total'] = df[['Math', 'Science', 'English']].sum(axis=1)

print("Total score for each student:")
print(df[['Name', 'Total']])
```

```python
# Create a column that compares English and Science scores
df['Better Subject'] = df.apply(lambda row: 'English' if row['English'] > row['Science'] else 'Science', axis=1)
#df.apply does the function on each row in the dataframe, lamda acts like an anonmyous function
'''
creating the function normally:
def better_subject(row):
    if row['English'] > row['Science']:
        return 'English'
    else:
        return 'Science'

# تطبيق الدالة على كل صف باستخدام apply()
df['Better Subject'] = df.apply(better_subject, axis=1)
'''
print("Comparison between English and Science scores:")
print(df[['Name', 'English', 'Science', 'Better Subject']])
```
